```python
import json
from transformers import pipeline

# GPT-2 modelini yükle
generator = pipeline('text-generation', model='gpt2')

def generate_content(news_item):
    """
    Haber başlığı ve açıklamasına dayalı olarak Türkçe içerik üretir.
    """
    prompt = (
        f"Aşağıdaki haber başlığı ve açıklamasına dayalı olarak Türkçe bir içerik oluşturun:\n\n"
        f"Başlık: {news_item['title']}\n"
        f"Açıklama: {news_item['description']}\n\n"
        f"İçerik:"
    )

    # Modelden içerik oluştur
    response = generator(prompt, max_new_tokens=100, num_return_sequences=1)
    content = response[0]['generated_text'].strip()

    return content

def generate_news_content(input_file, output_file, num_news=10):
    """
    Haber dosyasını yükler ve ilk 10 haber için Türkçe içerik oluşturur.
    """
    # Haberleri yükle
    with open(input_file, 'r', encoding='utf-8') as f:
        news_list = json.load(f)

    # İlk 10 haber için içerik oluştur
    generated_news = []
    for news_item in news_list[:num_news]:
        content = generate_content(news_item)
        generated_news.append({
            "title": news_item["title"],
            "description": news_item["description"],
            "content": content
        })

    # Oluşturulan içerikleri yeni bir JSON dosyasına kaydet
    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(generated_news, f, ensure_ascii=False, indent=4)

    print(f"İlk {num_news} haber için içerik oluşturuldu ve '{output_file}' dosyasına kaydedildi.")

# Örnek kullanım
input_file = 'tech_news.json'  # Giriş dosyası
output_file = 'icerik_haberler.json'  # Çıkış dosyası
generate_news_content(input_file, output_file)
```
